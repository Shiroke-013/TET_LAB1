```pyspark
#Seleccionando tres columnas: Edad, Estado y Sexo
df.select('Edad','Estado','Sexo').show(5)
```

```pyspark
#Imprimiendo primeras 5 filas de la nueva columna.
nuevo_data.select('ID de caso','Edad','Edad_en_10').show(5)
```

```pyspark
#Filtrando los datos para imprimir datos de personas con el sexo Masculino.
df.filter(df['Sexo']=='M').select('ID de caso','Edad','Sexo','Estado').show(5)
```

```pyspark
#Filtrando los datos para imprimir datos de personas que su estado sea diferente a Leve.
df.filter(df['Estado']!='Leve').select('ID de caso','Edad','Sexo','Estado').show(5)
```

```pyspark
#Cambiando nombre de las columnas
df2 = df.toDF('fecha reporte web', 'ID de caso', 'Fecha de notificacion', 'Codigo DIVIPOLA departamento', 'Nombre departamento', 'Codigo DIVIPOLA municipio', 'Nombre municipio', 'Edad', 'Unidad de medida de edad', 'Sexo', 'Tipo de contagio', 'Ubicacion del caso', 'Estado', 'Codigo ISO del pais', 'Nombre del pais', 'Recuperado', 'Fecha de inicio de sintomas', 'Fecha de muerte', 'Fecha de diagnostico', 'Fecha de recuperacion', 'Tipo de recuperacion', 'Pertenencia etnica', 'Nombre del grupo etnico')
df2.columns
```

```pyspark
# Seleccionando Columnas
df2.select('ID de caso','Edad','Sexo','Estado', 'Nombre del pais').show(5)
```

```pyspark
# Importando Tipos de Datos y UDF
from pyspark.sql.types import StringType,DoubleType,IntegerType
from pyspark.sql.functions import udf
```

```pyspark
# Función que mira si el paciente es Aleman o no.
def german_patient(country):
    if country in ['ALEMANIA']:
        return 'Aleman'
    else:
        return 'Extranjero'
```

```pyspark
#create udf using python function
german_udf=udf(german_patient,StringType())
#apply udf on dataframe
df2.withColumn('Ciudadano Aleman',german_udf(df2['Nombre del pais'])).show(5,False)
```

# Punto 3 y 4

```pyspark
#Los 10 departamentos con más casos de covid en Colombia ordenados de mayor a menor.
#df2.groupBy('Nombre departamento').count().orderBy('count',ascending=False).show(10,False)
df_dep = df2.groupBy('Nombre departamento').count().orderBy('count',ascending=False)
df_dep.coalesce(1).write.csv("s3://bigdatabucketadchavesp/Lab3-Parte3/Departamentos", header = True)
```

```pyspark
# Las 10 ciudades con más casos de covid en Colombia ordenados de mayor a menor.
#df2.groupBy('Nombre municipio').count().orderBy('count',ascending=False).show(10,False)
df_ciudades = df2.groupBy('Nombre municipio').count().orderBy('count',ascending=False)
df_ciudades.coalesce(1).write.csv("s3://bigdatabucketadchavesp/Lab3-Parte3/Ciudades", header = True)
```

```pyspark
# Los 10 días con más casos de covid en Colombia ordenados de mayor a menor.
#df2.groupBy('Fecha de diagnostico').count().orderBy('count',ascending=False).show(10,False)
df_dias = df2.groupBy('Fecha de diagnostico').count().orderBy('count',ascending=False)
df_dias.coalesce(1).write.csv("s3://bigdatabucketadchavesp/Lab3-Parte3/Dias", header = True)
```

```pyspark
# Distribución de casos por edades de covid en Colombia.
#df2.groupBy('Edad').count().orderBy('Edad',ascending=True).show(10,False)
df_edades = df2.groupBy('Edad').count().orderBy('Edad',ascending=True)
df_edades.coalesce(1).write.csv("s3://bigdatabucketadchavesp/Lab3-Parte3/Edades", header = True)
```

```pyspark
# Numero de Casos por Nacionalidad
#df2.groupBy('Nombre del Pais').count().orderBy('count',ascending=False).show(10,False)
df_nacionalidad = df2.groupBy('Nombre del Pais').count().orderBy('count',ascending=False)
df_nacionalidad.coalesce(1).write.csv("s3://bigdatabucketadchavesp/Lab3-Parte3/PorNacionalidad", header = True)
```
